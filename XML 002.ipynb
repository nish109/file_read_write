```python
#import os
import sys
import fileinput
import linecache
import numpy as np
import pandas as pd
import os.path
```

```python
#load in VoT data as dataframe
#a dataframe for each peak for VoT factors
amvotfile = pd.read_csv('am_vot.csv', header= 1) #these folowing csvs need to ve saved in the same folder as this code 
ipvotfile = pd.read_csv('ip_vot.csv', header= 1)
pmvotfile = pd.read_csv('pm_vot.csv', header= 1)
voc = pd.read_csv('voc.csv', header = 0)
am_vot = pd.DataFrame(amvotfile).set_index(['year'])
ip_vot = pd.DataFrame(ipvotfile).set_index(['year'])
pm_vot = pd.DataFrame(pmvotfile).set_index(['year'])
voc = pd.DataFrame(voc).set_index(['year'])
am_vot
```

```python
#open a base xml file to fully customise:
#changetime = AM, IP, PM
#changeyear = 2027, 2042, 2057
#changeyeardig = 27,42,57
#changegrowth = low, central, high
#hgvvot, lcvot,lnwvot,lwvot = to be changed based on the approprate cells in the am/ip/pm dataframes
#hgvvoc,lcvoc,lnwvoc,lwvoc = changed based on voc dataframe



year = ['2027','2042', '2057']
yeardig = ['27', '42', '57']
time = ['AM', 'IP', 'PM']

#change of 4 digit year in file contents and file name
for year,dig in zip(year, yeardig):
    original = open('changeyearchangetime_changegrowth.xml','r')  #file needs to be stored in same folder and level as this code
    data = original.read()
    original.close()
    data = data.replace('changeyear', year)
    data = data.replace('changedig', dig)
    new = open(year+'changetime_changegrowth.xml','w')
    new.write(data)
    new.close()
```

```python
#change of peak time in file contents and file name for each FY
time = ['AM', 'IP', 'PM']
for peak in time:
    file = open('2027changetime_changegrowth.xml', 'r')
    data = file.read()
    file.close()
    data = data.replace('changetime', peak)
    new = open('2027'+peak+'_changegrowth.xml','w')
    new.write(data)
    
for peak in time:
    file = open('2042changetime_changegrowth.xml', 'r')
    data = file.read()
    file.close()
    data = data.replace('changetime', peak)
    new = open('2042'+peak+'_changegrowth.xml','w')
    new.write(data)
    new.close()
    
for peak in time:
    file = open('2057changetime_changegrowth.xml', 'r')
    data = file.read()
    file.close()
    data = data.replace('changetime', peak)
    new = open('2057'+peak+'_changegrowth.xml','w')
    new.write(data)
    new.close()
```

```python
#change of vot factors for each peak based on FY 
year = ['2027','2042', '2057']
for year in year:
    file = open(year +'AM_changegrowth.xml', 'r+')
    data = file.read()
    year = int(year)
    data = data.replace('hgvvot', str(am_vot.loc[year]['hgv']))
    data = data.replace('lcvot', str(am_vot.loc[year]['commute']))
    data = data.replace('lnwvot', str(am_vot.loc[year]['nonwork']))
    data = data.replace('lwvot', str(am_vot.loc[year]['work']))
    data = data.replace('hgvvoc', str(voc.loc[year]['hgv']))
    data = data.replace('lcvoc', str(voc.loc[year]['commute']))
    data = data.replace('lnwvoc', str(voc.loc[year]['nonwork']))
    data = data.replace('lwvoc', str(voc.loc[year]['work']))
    file.seek(0)
    file.truncate()
    file.write(data)
    file.close()
    
year = ['2027','2042', '2057']
for year in year:
    file = open(year +'IP_changegrowth.xml', 'r+')
    data = file.read()
    year = int(year)
    data = data.replace('hgvvot', str(ip_vot.loc[year]['hgv']))
    data = data.replace('lcvot', str(ip_vot.loc[year]['commute']))
    data = data.replace('lnwvot', str(ip_vot.loc[year]['nonwork']))
    data = data.replace('lwvot', str(ip_vot.loc[year]['work']))
    data = data.replace('hgvvoc', str(voc.loc[year]['hgv']))
    data = data.replace('lcvoc', str(voc.loc[year]['commute']))
    data = data.replace('lnwvoc', str(voc.loc[year]['nonwork']))
    data = data.replace('lwvoc', str(voc.loc[year]['work']))
    file.seek(0)
    file.truncate()
    file.write(data)
    file.close()
    
year = ['2027','2042', '2057']
for year in year:
    file = open(year +'PM_changegrowth.xml', 'r+')
    data = file.read()
    year = int(year)
    data = data.replace('hgvvot', str(pm_vot.loc[year]['hgv']))
    data = data.replace('lcvot', str(pm_vot.loc[year]['commute']))
    data = data.replace('lnwvot', str(pm_vot.loc[year]['nonwork']))
    data = data.replace('lwvot', str(pm_vot.loc[year]['work']))  
    data = data.replace('hgvvoc', str(voc.loc[year]['hgv']))
    data = data.replace('lcvoc', str(voc.loc[year]['commute']))
    data = data.replace('lnwvoc', str(voc.loc[year]['nonwork']))
    data = data.replace('lwvoc', str(voc.loc[year]['work']))
    file.seek(0)
    file.truncate()
    file.write(data)
    file.close()
```

```python
pm = ['2027pm_Low.xml','2027pm_Central.xml','2027pm_High.xml','2042pm_Low.xml','2042pm_Central.xml','2042pm_High.xml','2057pm_Low.xml','2057pm_Central.xml','2057pm_High.xml']
hgv = []
lc = []
lnw = []
lw = []

for file in pm:
    #path = r"C:\Users\PAT103427\OneDrive - Mott MacDonald\Documents\Oilgate\XML" #change this if rerunning in a new location
    #completename = os.path.join(path, file)  
    f = open(file,'rt')
    filedata = f.read()
    hgv.append(linecache.getline(file, 273)[37:48])
    lc.append(linecache.getline(file, 294)[37:48])
    lnw.append(linecache.getline(file, 314)[37:48])
    lw.append(linecache.getline(file, 334)[37:48])
    f.close()
hgv = np.asarray(hgv).reshape(3,3)
lc = np.asarray(lc).reshape(3,3)
lnwv = np.asarray(lnw).reshape(3,3)
lw = np.asarray(lw).reshape(3,3)
pm_hgv = pd.DataFrame(hgv, columns=['Low','Central','High'], index=['hgv, 2027', 'hgv, 2042', 'hgv, 2057'])
pm_lc = pd.DataFrame(lc , columns=['Low','Central','High'], index=['lc, 2027', 'lc, 2042', 'lc, 2057'])
pm_lnw = pd.DataFrame(lnwv , columns=['Low','Central','High'], index=['lnw, 2027', 'lnw, 2042', 'lnw, 2057'])
pm_lw = pd.DataFrame(lw , columns=['Low','Central','High'], index=['lw, 2027', 'lw, 2042', 'lw, 2057'])
pm_hgv = pm_hgv.T #transpose df 
pm_lc = pm_lc.T
pm_lnw = pm_lnw.T
pm_lw = pm_lw.T
pm_vot = pd.concat([pm_hgv,pm_lc,pm_lnw,pm_lw], axis=1)
a = pm_vot.columns.str.split(', ', expand=True).values

#swap values in NaN and replace NAN to ''
pm_vot.columns = pd.MultiIndex.from_tuples([('', x[0]) if pd.isnull(x[1]) else x for x in a])
pm_vot
```
